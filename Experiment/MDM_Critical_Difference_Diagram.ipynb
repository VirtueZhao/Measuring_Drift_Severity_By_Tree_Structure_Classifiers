```python
Datasets = ['Adult', 'Bank', 'Cardio', 'Chess', 'Credit', 'Diamonds', 'Gamma', 'Poker']
MDMs = [
    [2, 1.5, 2], [2, 4, 3.5], [2, 1.5, 3],
    [1.5, 2, 1.5], [2, 2, 1.5], [2, 1.5, 2], 
    [2, 1.5, 2], [2, 2, 2]
    ]
Accs = [
    [1, 4, 2], [1, 1.5, 1.5], [1, 4, 1.5], 
    [1.5, 2, 4], [1, 2, 4], [1, 4, 2], 
    [1, 4, 2], [1, 2, 2]
    ]
WDs = [
    [4, 1.5, 2], [4, 1.5, 1.5], [4, 1.5, 1.5], 
    [4, 2, 1.5], [4, 2, 1.5], [4, 1.5, 2], 
    [4, 1.5, 2], [4, 2, 2]
    ]
PIs = [
    [3, 3, 4], [3, 3, 3.5], [3, 3, 4], 
    [3, 4, 3], [3, 4, 3], [3, 3, 4], 
    [3, 3, 4], [3, 4, 4]
    ]

for i in range(len(Datasets)):
    dataset = Datasets[i]
    MDM = MDMs[i]
    Acc = Accs[i]
    WD = WDs[i]
    PI = PIs[i]
#     print("Dataset: " + dataset)
#     print("MDM: " + str(MDM))
#     print("Acc: " + str(Acc))
#     print("WD: " + str(WD))
#     print("PI: " + str(PI))

    MDM_mean = round(mean(MDM),2)
    Acc_mean = round(mean(Acc),2)
    WD_mean = round(mean(WD),2)
    PI_mean = round(mean(PI),2)
    
    MDM_stdev = round(stdev(MDM),2)
    Acc_stdev = round(stdev(Acc), 2)
    WD_stdev = round(stdev(WD), 2)
    PI_stdev = round(stdev(PI), 2)
    
    MDM_name = "MDM(" + str(MDM_mean) + "\u00B1" + str(MDM_stdev) + ")"
    Acc_name = "Acc(" + str(Acc_mean) + "\u00B1" + str(Acc_stdev) + ")"
    WD_name = "WD(" + str(WD_mean) + "\u00B1" + str(WD_stdev) + ")"
    PI_name = "PI(" + str(PI_mean) + "\u00B1" + str(PI_stdev) + ")"
#     print(MDM_name)
#     print(Acc_name)
#     print(WD_name)
#     print(PI_name)
    
    names = [MDM_name, Acc_name, WD_name, PI_name]
    avranks =  [MDM_mean, Acc_mean, WD_mean, PI_mean]
    cd = Orange.evaluation.compute_CD(avranks, 3, alpha='0.1') #tested on 3 datasets
    file_name = "MDM_CD_" + dataset + ".pdf"
    print(file_name)
#     CriticalDifferenceGraph.graph_ranks(avranks, names, cd=cd, width=5, textspace=1.5, filename=file_name)
    CriticalDifferenceGraph.graph_ranks(avranks, names, cd=cd, width=5, textspace=1.5)
```

```python
Drift_Types = ['Conditional Drift', 'Marginal Drift for Majority Class', 'Marginal Drift for Minority Class']
MDMs = [
    [2, 2, 2, 1.5, 2, 2, 2, 2],
    [1.5, 4, 1.5, 2, 2, 1.5, 1.5, 2],
    [2, 3.5, 3, 1.5, 1.5, 2, 2, 2]
    ]
Accs = [
    [1, 1, 1, 1.5, 1, 1, 1, 1],
    [4, 1.5, 4, 2, 2, 4, 4, 2],
    [2, 1.5, 1.5, 4, 4, 2, 2, 2]
]
WDs = [
    [4, 4, 4, 4, 4, 4, 4, 4],
    [1.5, 1.5, 1.5, 2, 2, 1.5, 1.5, 2],
    [2, 1.5, 1.5, 1.5, 1.5, 2, 2, 2]
]
PIs = [
    [3, 3, 3, 3, 3, 3, 3, 3],
    [3, 3, 3, 4, 4, 3, 3, 4],
    [4, 3.5, 4, 3, 3, 4, 4, 4]
]

for i in range(len(Drift_Types)):
    Drift_Type = Drift_Types[i]
    MDM = MDMs[i]
    Acc = Accs[i]
    WD = WDs[i]
    PI = PIs[i]
    print("Drift Type: " + Drift_Type)
    print("MDM: " + str(MDM))
    print("Acc: " + str(Acc))
    print("WD: " + str(WD))
    print("PI: " + str(PI))

    MDM_mean = round(mean(MDM),2)
    Acc_mean = round(mean(Acc),2)
    WD_mean = round(mean(WD),2)
    PI_mean = round(mean(PI),2)
    
    MDM_stdev = round(stdev(MDM),2)
    Acc_stdev = round(stdev(Acc), 2)
    WD_stdev = round(stdev(WD), 2)
    PI_stdev = round(stdev(PI), 2)
    
    
    MDM_mean_str = "{:.2f}". format(MDM_mean)
    MDM_stdev_str = "{:.2f}". format(MDM_stdev)
    MDM_name = "MDM(" + MDM_mean_str + "\u00B1" + MDM_stdev_str + ")"
    
    Acc_mean_str = "{:.2f}". format(Acc_mean)
    Acc_stdev_str = "{:.2f}". format(Acc_stdev)
    Acc_name = "Acc(" + Acc_mean_str + "\u00B1" + Acc_stdev_str + ")"
    
    WD_mean_str = "{:.2f}". format(WD_mean)
    WD_stdev_str = "{:.2f}". format(WD_stdev)
    WD_name = "WD(" + WD_mean_str + "\u00B1" + WD_stdev_str + ")"
    
    PI_mean_str = "{:.2f}". format(PI_mean)
    PI_stdev_str = "{:.2f}". format(PI_stdev)
    PI_name = "PI(" + PI_mean_str + "\u00B1" + PI_stdev_str + ")"
    print(MDM_name)
    print(Acc_name)
    print(WD_name)
    print(PI_name)
    
    names = [MDM_name, Acc_name, WD_name, PI_name]
    avranks =  [MDM_mean, Acc_mean, WD_mean, PI_mean]
    cd = Orange.evaluation.compute_CD(avranks, 3, alpha='0.1', test='bonferroni-dunn') #tested on 3 datasets
    file_name = "MDM_CD_" + Drift_Type + ".pdf"
    print(file_name)
    CriticalDifferenceGraph.graph_ranks(avranks, names, cd=cd, width=5, textspace=1.5, filename=file_name)
#     CriticalDifferenceGraph.graph_ranks(avranks, names, cd=cd, width=5, textspace=1.5)
```
